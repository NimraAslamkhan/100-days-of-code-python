```python
#PASSWORD GENERATOR
import random

letters = ["a", "b", "c", "d"]
numbers = ["1", "2", "3"]
symbols = ["!", "@", "#"]

password = ""

for _ in range(4):
    password += random.choice(letters)

for _ in range(2):
    password += random.choice(numbers)

for _ in range(2):
    password += random.choice(symbols)

print(password)
```

```python
#HANGMAN GAME
import random

words = ["apple", "banana", "mango"]
chosen_word = random.choice(words)

display = ["_"] * len(chosen_word)

if "_" not in display:
    print("You win!")


lives = 6

guess = input("Guess letter: ")

if guess not in chosen_word:
    lives -= 1
    print("Wrong guess")

print("Lives left:", lives)

while "_" in display and lives > 0:
    guess = input("Guess: ")

    for i in range(len(chosen_word)):
        if chosen_word[i] == guess:
            display[i] = guess

    if guess not in chosen_word:
        lives -= 1

    print(display)

if lives == 0:
    print("You lose")
else:
    print("You win")
```

```python
#GRADING PROGRAM

marks = {"Ali": 85, "Sara": 92}

grades = {}

for student in marks:
    score = marks[student]

    if score > 90:
        grades[student] = "A"
    elif score > 80:
        grades[student] = "B"
    else:
        grades[student] = "C"

print(grades)
```

```python
#SECRET AUCTION PROGRAM

bids = {}

while True:
    name = input("Enter name: ")
    bid = int(input("Enter bid: "))

    bids[name] = bid

    more = input("More bidders? yes/no: ")
    if more == "no":
        break

winner = max(bids, key=bids.get)

print(f"Winner is {winner}")
```

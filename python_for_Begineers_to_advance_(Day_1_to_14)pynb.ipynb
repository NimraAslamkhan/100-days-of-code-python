# hangman game

```python
# hangman game
import random

# -----------------------------
# WORD LIST
# -----------------------------
words = ["apple", "tiger", "banana", "car", "python", "school", "friend", "garden", "camera"]

# -----------------------------
# PICK A RANDOM WORD
# -----------------------------
word = random.choice(words)
word_letters = list(word)
blanks = ["_"] * len(word)

attempts = 6
guessed_letters = []

print("🎉 Welcome to Hangman Game!")
print("Guess the word letter by letter.")
print("You have", attempts, "attempts.\n")

# -----------------------------
# MAIN GAME LOOP
# -----------------------------
while attempts > 0 and "_" in blanks:

    print("Word:", " ".join(blanks))
    print("Attempts left:", attempts)
    print("Guessed letters:", guessed_letters)

    guess = input("\nEnter a letter: ").lower()

    # Check input validity
    if len(guess) != 1 or not guess.isalpha():
        print("❌ Please enter ONE alphabet letter only.\n")
        continue

    if guess in guessed_letters:
        print("⚠️ You already guessed that letter!\n")
        continue

    guessed_letters.append(guess)

    # -----------------------------
    # CHECK IF GUESS IS CORRECT
    # -----------------------------
    if guess in word_letters:
        print("✔ Correct guess!\n")

        # Reveal the guessed letter in the blanks
        for i in range(len(word)):
            if word[i] == guess:
                blanks[i] = guess
    else:
        attempts -= 1
        print("❌ Wrong guess! Attempts left:", attempts, "\n")

# -----------------------------
# END OF GAME
# -----------------------------
if "_" not in blanks:
    print("🎉 CONGRATULATIONS! You guessed the word:", word)
else:
    print("💀 GAME OVER! The correct word was:", word)
```

# caeser cipher game by using functions

```python
# Simple Caesar Cipher Game

# -----------------------------
# Ask user for choice
# -----------------------------
print("Welcome to Caesar Cipher Game!")
print("1. Encrypt a message")
print("2. Decrypt a message")

choice = input("Enter 1 or 2: ")

# -----------------------------
# Ask user for message and shift
# -----------------------------
message = input("Enter your message: ")
shift = int(input("Enter shift number (e.g., 3): "))

# -----------------------------
# Function to encrypt
# -----------------------------
def encrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():  # Check if character is a letter
            # Shift lowercase letters
            if char.islower():
                result += chr((ord(char) - 97 + shift) % 26 + 97)
            # Shift uppercase letters
            else:
                result += chr((ord(char) - 65 + shift) % 26 + 65)
        else:
            result += char  # Keep numbers, symbols, spaces as is
    return result

# -----------------------------
# Function to encrypt
# -----------------------------
def decrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            if char.islower():
                result += chr((ord(char) - 97 - shift) % 26 + 97)
            else:
                result += chr((ord(char) - 65 - shift) % 26 + 65)
        else:
            result += char
    return result

# -----------------------------
# Main logic
# -----------------------------
if choice == "1":
    print("Encrypted message:", encrypt(message, shift))
elif choice == "2":
    print("Decrypted message:", decrypt(message, shift))
else:
    print("Invalid choice! Please enter 1 or 2.")
```

```python
# advance Caesar Cipher Game

def encrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            if char.isupper():
                # Uppercase letters
                result += chr((ord(char) - 65 + shift) % 26 + 65)
            else:
                # Lowercase letters
                result += chr((ord(char) - 97 + shift) % 26 + 97)
        else:
            result += char  # keep symbols/numbers/spaces
    return result

def decrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            if char.isupper():
                result += chr((ord(char) - 65 - shift) % 26 + 65)
            else:
                result += chr((ord(char) - 97 - shift) % 26 + 97)
        else:
            result += char
    return result

# -----------------------------
# Main Program
# -----------------------------
print("🎯 Caesar Cipher Game")
print("1. Encrypt a message")
print("2. Decrypt a message")

choice = input("Enter 1 or 2: ")
message = input("Enter your message: ")
shift = int(input("Enter shift number (e.g., 3): "))

if choice == "1":
    encrypted = encrypt(message, shift)
    print("Encrypted message:", encrypted)
elif choice == "2":
    decrypted = decrypt(message, shift)
    print("Decrypted message:", decrypted)
else:
    print("❌ Invalid choice")
```

# Black Jeck Game

```python
import random

# Deal one random card
def deal_card():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]  # 11 = Ace
    return random.choice(cards)

# Calculate score
def calculate_score(cards):
    # Check for Blackjack (Ace + 10)
    if sum(cards) == 21 and len(cards) == 2:
        return 0

    # Convert Ace (11) to 1 if score > 21
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)

    return sum(cards)

# Compare scores
def compare(user_score, computer_score):
    if user_score == computer_score:
        return "Draw 😐"
    elif computer_score == 0:
        return "You lose! Opponent has Blackjack 😭"
    elif user_score == 0:
        return "You win with a Blackjack! 🎉"
    elif user_score > 21:
        return "You went over 21. You lose 😢"
    elif computer_score > 21:
        return "Opponent went over 21. You win 🎉"
    elif user_score > computer_score:
        return "You win 🎉"
    else:
        return "You lose 😢"

def play_game():
    user_cards = []
    computer_cards = []
    is_game_over = False

    # Starting two cards each
    for _ in range(2):
        user_cards.append(deal_card())
        computer_cards.append(deal_card())

    # Game loop
    while not is_game_over:
        user_score = calculate_score(user_cards)
        computer_score = calculate_score(computer_cards)

        print(f"Your cards: {user_cards}, current score: {user_score}")
        print(f"Computer's first card: {computer_cards[0]}")

        # If Blackjack or bust
        if user_score == 0 or computer_score == 0 or user_score > 21:
            is_game_over = True
        else:
            should_continue = input("Type 'y' to get another card, 'n' to pass: ")

            if should_continue == "y":
                user_cards.append(deal_card())
            else:
                is_game_over = True

    # Computer draws until score >= 17 (and not Blackjack)
    while computer_score != 0 and computer_score < 17:
        computer_cards.append(deal_card())
        computer_score = calculate_score(computer_cards)

    print(f"\nYour final hand: {user_cards}, final score: {user_score}")
    print(f"Computer's final hand: {computer_cards}, final score: {computer_score}")
    print(compare(user_score, computer_score))


# Start the game
play_game()
```

# Number Genrating Game

```python
from random import randint

EASY_LEVEL_TURNS = 10
HARD_LEVEL_TURNS = 5

def check_answer(user_guess, actual_answer, turns):
    if user_guess > actual_answer:
        print("**Too high**")
        return turns - 1
    elif user_guess < actual_answer:
        print("**Too low**")
        return turns - 1
    else:
        print(f"You got it! The answer was {actual_answer}.")
        return turns

def set_difficulty():
    level = input('Choose a difficulty: type "easy" or "hard": ')
    if level.lower() == "easy":
        return EASY_LEVEL_TURNS
    else:
        return HARD_LEVEL_TURNS

def game():
    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number between 1 and 100.")

    answer = randint(1, 100)
    print(f"Pssst... the correct answer is {answer}")  # remove this in real game

    turns = set_difficulty()

    guess = 0
    while guess != answer:
        print(f"You have {turns} attempts remaining to guess the number.")
        guess = int(input("Make a guess: "))

        turns = check_answer(guess, answer, turns)

        if turns == 0:
            print("You've run out of guesses. You lose!")
            return
        elif guess != answer:
            print("Guess again!")

game()

```

```python
# -------------------------------
#  Class: Course
# -------------------------------
class Course:
    def __init__(self, course_name, credit_hours):
        self.course_name = course_name
        self.credit_hours = credit_hours


# -------------------------------
#  Class: Student
# -------------------------------
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.__marks = {}   # Encapsulated (Private)

    def enroll(self, course):
        self.__marks[course.course_name] = None

    def add_marks(self, course_name, marks):
        if course_name in self.__marks:
            self.__marks[course_name] = marks

    def calculate_grade(self, marks):
        if marks >= 90:
            return "A"
        elif marks >= 80:
            return "B"
        elif marks >= 70:
            return "C"
        elif marks >= 60:
            return "D"
        else:
            return "F"

    def get_report(self):
        print(f"\nStudent: {self.name} (ID: {self.student_id})")
        print("Course Report:")
        for course, marks in self.__marks.items():
            if marks is None:
                print(f" - {course}: Not Graded yet")
            else:
                grade = self.calculate_grade(marks)
                print(f" - {course}: {marks} → Grade: {grade}")


# -------------------------------
#  Inheritance Example
# -------------------------------
class ScholarshipStudent(Student):
    def __init__(self, name, student_id, scholarship_percent):
        super().__init__(name, student_id)
        self.scholarship_percent = scholarship_percent

    # Polymorphism (different fee structure)
    def get_fee(self, base_fee):
        discounted = base_fee - (base_fee * self.scholarship_percent / 100)
        return discounted


# -------------------------------
#  School Class (Manager)
# -------------------------------
class School:
    def __init__(self):
        self.students = []
        self.courses = []

    def add_student(self, student):
        self.students.append(student)

    def add_course(self, course):
        self.courses.append(course)

    def show_all_students(self):
        print("\nAll Students:")
        for stu in self.students:
            print(f"- {stu.name} ({stu.student_id})")


# -------------------------------
# Usage Example
# -------------------------------
school = School()

# Create courses
math = Course("Mathematics", 3)
python = Course("Python Programming", 4)

school.add_course(math)
school.add_course(python)

# Create students
s1 = Student("Nimra", "ST001")
s2 = ScholarshipStudent("Ayesha", "ST002", scholarship_percent=50)

school.add_student(s1)
school.add_student(s2)

# Enroll students
s1.enroll(math)
s1.enroll(python)

s2.enroll(math)

# Add marks
s1.add_marks("Mathematics", 92)
s1.add_marks("Python Programming", 88)

s2.add_marks("Mathematics", 75)

# Show Reports
s1.get_report()
s2.get_report()

# Scholarship fee example
print("\nFee Calculation for Scholarship Student:")
print("Original Fee: 40,000")
print("Discounted Fee:", s2.get_fee(40000))
```

```python
## sanke game
import turtle
import time
import random

# -------------------------------
# Setup Screen
# -------------------------------
screen = turtle.Screen()
screen.title("Snake Game")
screen.bgcolor("black")
screen.setup(width=600, height=600)
screen.tracer(0)  # turns off auto screen updates

# -------------------------------
# Create Snake Head
# -------------------------------
head = turtle.Turtle()
head.shape("square")
head.color("green")
head.penup()
head.goto(0, 0)
head.direction = "stop"

# -------------------------------
# Create Food
# -------------------------------
food = turtle.Turtle()
food.shape("circle")
food.color("red")
food.penup()
food.goto(0, 100)

# -------------------------------
# Snake Body Segments
# -------------------------------
segments = []

# -------------------------------
# Score
# -------------------------------
score = 0
high_score = 0

pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0, 260)
pen.write("Score: 0  High Score: 0", align="center", font=("Courier", 24, "normal"))

# -------------------------------
# Functions
# -------------------------------
def go_up():
    if head.direction != "down":
        head.direction = "up"

def go_down():
    if head.direction != "up":
        head.direction = "down"

def go_left():
    if head.direction != "right":
        head.direction = "left"

def go_right():
    if head.direction != "left":
        head.direction = "right"

def move():
    if head.direction == "up":
        y = head.ycor()
        head.sety(y + 20)
    if head.direction == "down":
        y = head.ycor()
        head.sety(y - 20)
    if head.direction == "left":
        x = head.xcor()
        head.setx(x - 20)
    if head.direction == "right":
        x = head.xcor()
        head.setx(x + 20)

# -------------------------------
# Keyboard bindings
# -------------------------------
screen.listen()
screen.onkeypress(go_up, "Up")
screen.onkeypress(go_down, "Down")
screen.onkeypress(go_left, "Left")
screen.onkeypress(go_right, "Right")

# -------------------------------
# Main Game Loop
# -------------------------------
while True:
    screen.update()

    # Check collision with wall
    if head.xcor() > 290 or head.xcor() < -290 or head.ycor() > 290 or head.ycor() < -290:
        time.sleep(1)
        head.goto(0, 0)
        head.direction = "stop"

        # Hide segments
        for segment in segments:
            segment.goto(1000, 1000)
        segments.clear()

        score = 0
        pen.clear()
        pen.write(f"Score: {score}  High Score: {high_score}", align="center", font=("Courier", 24, "normal"))

    # Check collision with food
    if head.distance(food) < 20:
        # Move food to random spot
        x = random.randint(-14, 14) * 20
        y = random.randint(-14, 14) * 20
        food.goto(x, y)

        # Add segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("green")
        new_segment.penup()
        segments.append(new_segment)

        # Increase score
        score += 10
        if score > high_score:
            high_score = score
        pen.clear()
        pen.write(f"Score: {score}  High Score: {high_score}", align="center", font=("Courier", 24, "normal"))

    # Move body segments
    for index in range(len(segments)-1, 0, -1):
        x = segments[index-1].xcor()
        y = segments[index-1].ycor()
        segments[index].goto(x, y)

    if len(segments) > 0:
        segments[0].goto(head.xcor(), head.ycor())

    move()

    # Check collision with body
    for segment in segments:
        if segment.distance(head) < 20:
            time.sleep(1)
            head.goto(0, 0)
            head.direction = "stop"
            for segment in segments:
                segment.goto(1000, 1000)
            segments.clear()
            score = 0
            pen.clear()
            pen.write(f"Score: {score}  High Score: {high_score}", align="center", font=("Courier", 24, "normal"))

    time.sleep(0.1)
```

# File handling


### How to Open, Read, and Write to Files using the "with" Keyword
Understand Relative and Absolute File Paths
Understanding Relative and Absolute File Paths (Beginner-Friendly)

When working with files in Python (or any programming language), you must tell the program where a file is located. This location is called a file path.

### There are two types of file paths:

Absolute Path

Relative Path

### Absolute File Path ✅ What is it?

An absolute path gives the full location of a file, starting from the root directory of the system.

➡️ It does not depend on where your program is running.

Think of it like:

Giving someone your complete home address including city, street, and house number.

🐍 Python Example: file = open("C:/Users/HP/Documents/project/data.txt", "r")

✅ Advantages:

Always points to the correct file

No confusion

❌ Disadvantages:

Not portable (won’t work on another computer)

Long and hard to maintain

### Relative File Path ✅ What is it?

A relative path gives the file location relative to the current working directory (where your Python file is running).

➡️ It depends on where your script is located.

📁 Example Folder Structure: project/ │ ├── main.py ├── data/ │ └── file.txt

📌 Relative Path from main.py: file = open("data/file.txt", "r")

🧠 Think of it like:

Saying “go two steps forward and turn left” instead of giving a full address.

🔼 Special Relative Symbols Symbol Meaning . Current directory .. Parent directory Example: open("./data/file.txt") # same folder → data open("../file.txt") # one folder up

🆚 Absolute vs Relative (Quick Comparison) Feature Absolute Path Relative Path Starts from root ✅ Yes ❌ No Depends on location ❌ No ✅ Yes Portable ❌ No ✅ Yes Easy to maintain ❌ No ✅ Yes ✅ Best Practice (Very Important ⭐)

For projects, always use relative paths.

✔️ Works on any system ✔️ Clean and professional ✔️ Used in real-world projects

🐍 Python Tip: Check Current Working Directory import os print(os.getcwd())

This shows where Python is currently running, which is important for relative paths.

🔥 Simple Rule to Remember

Absolute Path → Full address 📍

Relative Path → From where you are now 🚶‍♂️
